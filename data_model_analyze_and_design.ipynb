```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/companydata.csv", encoding='cp949')
```

```python
print(df)
```

```python
total_revenue = df.iloc[0][1:] #총매출
print(total_revenue)
```

```python
target_sales = df.iloc[1][1:] #목표매출
advertising_cost = df.iloc[2][1:] #광고비용
social_network_cost  = df.iloc[3][1:] #소셜네트워크비용
```

```python
bar = plt.bar(["Sep","Oct","Nov","Dec","Jan","Feb"],[5280000, 5500000, 5729000,5968000, 6217000,6476000 ])


plt.title("Monthly target sales")
plt.xlabel("month")
plt.ylabel("tagret sales")

for rect in bar:
  height = rect.get_height()
  plt.text(rect.get_x() + rect.get_width()/2.0, height, height,ha='center', va='bottom', size = 8)

plt.show()
```

```python
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],total_revenue,label='total_revenue')
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],target_sales,label='target_sales')

plt.title('total revenue and target sales')
plt.xlabel('Month')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 2))
plt.subplot(1, 2, 1)
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"], advertising_cost,color='r', label='Advertising Cost')
plt.xlabel('Month')
plt.ylabel('Cost')
plt.title('Advertising Cost')
plt.gca().invert_yaxis()
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"], total_revenue, label='total revenue')
plt.xlabel('Month')
plt.ylabel('Cost')
plt.title('total revenue')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 2))
plt.subplot(1, 2, 1)
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"],social_network_cost, color='g',label='Social Network Cost')
plt.xlabel('Month')
plt.ylabel('Cost')
plt.title('Social Network Cost')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(["Sep","Oct","Nov","Dec","Jan","Feb"], total_revenue, label='total revenue')
plt.xlabel('Month')
plt.ylabel('Cost')
plt.title('total revenue')
plt.legend()
plt.show()
```
